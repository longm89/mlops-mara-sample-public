```python
def statistic_num(df, num_cols):
    a = df[num_cols].astype(float).isna().agg(['sum', 'mean']).T
    a = a.rename(columns={'sum': 'Missing', 'mean': '% Missing'}).T
    b = df[num_cols].astype(float).describe(include = "all")
    res = pd.concat([a,b]).T
    return res
```

```python
def statistic_cat(df, cat_cols):
    a = df[cat_cols].astype(str).isna().agg(['sum', 'mean']).T
    a = a.rename(columns={'sum': 'Missing', 'mean': '% Missing'}).T
    b = df[cat_cols].astype(str).describe(include = "all")
    res = pd.concat([a,b]).T
    return res
```
